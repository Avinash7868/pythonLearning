
### Assignment 14: Counting with Dictionaries

Write a function that takes a string and returns a dictionary with the count of each character in the string. Print the dictionary.


```python
def count_characters(strr):
    count_dict = {}
    for chars in strr:
        count_dict[chars] = count_dict.get(chars,0) + 1
    print(count_dict)

count_characters("hello world this is Avinash ojha")
```

```python
def count_word_frequency(sentence):
    # Your code goes here
    print(len(sentence))
    if not sentence:
        return {}
    count = {}
    split_word = sentence.split(" ")
    for x in split_word:
        count[x] = count.get(x,0) + 1
    return count

sentence = "hello world hello"
print(count_word_frequency(sentence))
print(count_word_frequency(""))
```

```python
def merge_dicts_with_overlapping_keys(dicts):
    # Your code goes here
    if len(dicts) <= 0 :
        return {}
    res = {}
    for x in dicts:
        if isinstance(x,dict):
            for key,value in x.items():
                res[key] = res.get(key,0) + value
    return res

input1 = [{'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'c': 5, 'd': 6}]
input2  = [{'x': 10, 'y': 20}, {'y': 30, 'z': 40}, {'z': 50, 'x': 60}]

print(merge_dicts_with_overlapping_keys(input1))
print(merge_dicts_with_overlapping_keys(input2))
```
