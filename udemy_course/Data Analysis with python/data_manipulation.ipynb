```python
# Handling Missing values
df.isnull()
```

```python
# I have to make a Nan data to learn how to handle empty or Nan data
df['Total Cost'][0] = None
```

```python
df['Total Cost'][0]
```

```python
# To check null value by column
df.isnull().any(axis=1)
# To check null value by row
# df.isnull().any(axis=0)
```

```python
df.isnull().sum()
```

```python
# To fill the na/null data with some other value (not good apporoach)
df_filled = df.fillna(0) 
df_filled.head(2)
```

```python
# Filling missing value with the mean of the columne
df['Total_Cost_fillna'] = df['Total Cost'].fillna(df['Total Cost'].mean())
df.head(2)
```

```python
df.dtypes
# Object is basically a string
```

```python
# Renaming the column
df = df.rename(columns={'Ship Date':"Shipped Date"})
df.head(2)
```

```python
# Change the data type
df["order_id_float"] = df["Order ID"].astype(float)
df.head(2)
```

```python
# Apply a function to any column to perform any operations or changes
df['New_Unit_Cost']= df["Unit Cost"].apply(lambda x: x*2)
df.head(2)
```

```python
# Datat aggregating and grouping
grouped_mean = df.groupby('Region')['Total Profit'].mean()
print("grouped_mean of Total Profit based on Region \n", grouped_mean)
```

```python
# Grouping 2 different column with total profit
grouped_sum = df.groupby(['Region',"Country"])['Total Profit'].sum()
print(grouped_sum)
```

```python
# Using multiple aggregation function in a sigle time
grouped_data = df.groupby(['Region'])["Total Profit"].agg(['sum','mean','count'])
print(grouped_data)
```

```python
# Merging and joining data frames
df1 = pd.DataFrame({"key":["A","B","C"],"value1":[2,3,4]})
df2 = pd.DataFrame({"key":["A","B","D"],"value2":[7,8,9]})
```

```python
df1
```

```python
df2
```

```python
# Merge the data frame on the "key" colume
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```
