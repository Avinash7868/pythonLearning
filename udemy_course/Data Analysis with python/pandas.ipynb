```python
import pandas as pd

data  = [1,2,3,4,5]
series = pd.Series(data) 
print(series)
print(type(series))
```

```python
# Making series using dictionary
data= {"a":1,"b":2,"c":4}
print(data)
series = pd.Series(data)
print(series)
```

```python
data = [1,2,3,4]
index=["a","b","c","d"]
pd.Series(data,index=index)
```

```python
# Dataframe
# create a dataframe with dictonary
data = {
    "Name":['Avi',"john",'jack'],
    "Age":[22,33,44],
    "City":['Indore','New York','Dhar']
}
df = pd.DataFrame(data)
print(type(df))
df
```

```python
import numpy as np
# np.array(data)
np.array(df)
```
