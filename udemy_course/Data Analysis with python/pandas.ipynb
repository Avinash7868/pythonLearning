```python
# Data Manupulation with Data Frame , I am picking an easy data to manupulat instead of whole csv
data = [
    {"Name":"Avi",	"Age":22,"City":"Indore"},
    {"Name":"john",	"Age":22,"City":"New York"},
    {"Name":"jack",	"Age":22,"City":"Dhar"},
]
df2 = pd.DataFrame(data)
df2
```

```python
# Adding a column
df2['salary'] = [500000,80000,30000]
df2
```

```python
# Removing a column
# It wil give an error because by default drop will check row index because axis=0 is default 
# df2.drop('salary')

# axis = 1 means to check column instead of row
df2.drop('salary',axis=1)
```

```python
df2
# Salary is still there even after using drop
```

```python
# to permanently drop the data
df2.drop('salary',axis=1,inplace=True)
```

```python
df2
```

```python
# Add age to the column
df2['Age'] = df2['Age'] + 1
# df2['Age'] +=  1
```

```python
df2
```

```python
# Drop based on row
df2.drop(2)
```
