```python
import seaborn as sns
```

```python
# Basic ploting with seaborne
tips_data =sns.load_dataset('tips')
tips_data
```

```python
# scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips_data)
plt.title("Total bill VS tips")
plt.show()
```

```python
# Line plot
sns.lineplot(x='size',y='total_bill',data=tips_data)
plt.show()
```

```python
# Catogarical plot
sns.barplot(x="day",y="total_bill",data=tips_data)
plt.title("Total bill based on day")
plt.show()
```

```python
# Box plot
sns.boxplot(x="day",y="total_bill",data=tips_data)
plt.title("Box plot based on day and total bill")
plt.show()
```

```python
# violin plot
sns.violinplot(x="day",y="total_bill",data=tips_data)
plt.show()
```

```python
# Histogram
sns.histplot(tips_data['total_bill'],bins=10,kde=True)
```

```python
# KDE plot
sns.kdeplot(tips_data['total_bill'],fill=True)
```

```python
# Pair Plot
sns.pairplot(tips_data)
plt.show()
```

```python
# Heat map
corr = tips_data[['total_bill',"tip",'size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
